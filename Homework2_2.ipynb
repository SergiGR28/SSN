```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
#Definicio de variables SI
hred_SI = 1.0545718e-34 #Constant de Planck reduida en J·s
m_e_SI = 9.10938356e-31 #Masa del electro en Kg
e_SI = 1.602176634e-19 #Carrega del electro en C
epsilon_0_SI = 8.854187817e-12 #Constant de permitivitat en F/m
a0_SI = 4 * np.pi * epsilon_0_SI * hred_SI**2 / (m_e_SI * e_SI**2) #Radi de bohr en m
```

```python
#Equacio de Schrödinguer (SI)
r = sp.symbols('r')
psi = sp.Function('psi')(r)
V_SI = -e_SI**2 / (4 * np.pi * epsilon_0_SI * r)
H_SI = -hred_SI**2 / (2 * m_e_SI * r**2) * sp.diff(psi, r, 2) + V_SI * psi
print('Equació de Schrödinger:')
sp.pprint(H_SI, use_unicode=True)
```

```python
#Equació de Schrödinguer (UA)
hred_UA = 1 #Constant de Planck reduida en unitats atomiques
m_e_UA = 1 #Massa del electró en unitats atomiques
e_UA = 1 #Carrega del electró en unitat atomiques
V_UA = -1/r
H_UA = (-hred_UA**2 / (2 * m_e_UA)) * sp.diff(psi, r, 2) + V_UA * psi
print('\nEquació de Schröndinger en UA:')
sp.pprint(H_UA, use_unicode=True)
```

```python
#Solució de la funció d'ona fonamental (estat 1s)
def psi_1s(r):
  return (1 / np.sqrt(np.pi)) * np.exp(-r)
```

```python
#Grafica de la funció d'ona fonamental
r_values = np.linspace(0, 5, 100)
psi_values = psi_1s(r_values)

plt.figure(figsize=(8, 5))
plt.plot(r_values, psi_values, label=r'$\psi_{1s}(r)$')
plt.xlabel('Distancia r (en radis de Bohr)')
plt.ylabel('Funció d\'ona')
plt.title('Funció d\'ona fonamental del atom d\'hidrogen')
plt.legend()
plt.grid()
plt.show()
```
